```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(pd.read_csv('region_click_percentage.csv'))
```

We can create a new dataframe to sum the local clicks for jobs by region to see how each region gets local attention for jobs in comparison to outsourced attention. 

```python
total_local_clicks = pd.DataFrame(None)
total_local_clicks["Local_click_sum"] = df.groupby(['Region'])['localClicks'].sum()
total_local_clicks = total_local_clicks.reset_index()
df = df.merge(total_local_clicks, on = "Region")
df
```

```python
df["local_click_sum/clicks_y"] = df["localClicks"]/df["Local_click_sum"]
df["local_click_sum/clicks_y"].sum()
```

```python
df["all_total_clicks"] = df["clicks_x"].sum()
local_clicks_by_edu = pd.DataFrame(None)
local_clicks_by_edu = df.groupby(["educationRequirements"])["localClicks"].sum().reset_index()
local_clicks_by_edu
```

```python
df = df.merge(local_clicks_by_edu, on=['educationRequirements'])
df
```

```python
df['Local_by_edu/Total'] = df['localClicks_y']/df['all_total_clicks']
```

```python
df_sub = df[["educationRequirements", "Local_by_edu/Total"]]
df_sub = df_sub.drop_duplicates()
df_sub.to_csv('Local_clicks_by_edu_over_total.csv')
```
